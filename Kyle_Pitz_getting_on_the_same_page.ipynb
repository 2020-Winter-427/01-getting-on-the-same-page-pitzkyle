```python
import matplotlib.pyplot as plt
import numpy as np

## Define mean and variance for the normal distribution and plot the histogram
mu, sigma = 0, 1
data_norm = np.random.normal(mu, sigma, 1000)
hist_norm = plt.hist(data_norm, 25, density=True)

## Create a normal probability density funtion to compare with the histogram, add labels
import scipy.stats as sp
x = np.linspace(-3, 3, 1000)
pdf_norm = sp.norm.pdf(x)
plt.plot(x,pdf_norm)
plt.title('Random normal histogram')
plt.xlabel('x value bins')
plt.ylabel('random data value counts (normalized)')
plt.show()

## Create a scatter plot of the normal distribution data
plt.plot(data_norm, '.')
plt.title('Normal scatter plot')
plt.xlabel('data point')
plt.ylabel('random data value')
plt.show()
# We can see here that a lot of the data points surrounding zero begins to merge together,
# making the individual points hard to distinguish.




## Create a random exponential distribution and plot the histogram
data_exp = np.random.exponential(1, 1000)
hist_exp = plt.hist(data_exp, 25, density=True)

## Create a exponential probability density function to compare with the histogram, add labels
y = np.linspace(0, 5, 1000)
pdf_exp = sp.expon.pdf(y)
plt.plot(y, pdf_exp)
plt.title('Random exponential histogram')
plt.xlabel('x value bins')
plt.ylabel('random data value counts (normalized)')
plt.show()

## Create a scatter plot of the exponential distribution data
plt.plot(data_exp, '.')
plt.title('Exponential scatter plot')
plt.xlabel('data point')
plt.ylabel('random data value')
plt.show()
# With this plot, any of the data nearest zero is incredibly blurry to look at, 
# and really begins to lump together.


```
